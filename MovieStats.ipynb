```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
%matplotlib inline
import json
import math
```

```python
r = requests.get("http://www.the-numbers.com/movie/records/All-Time-Domestic-Box-Office")
soup = BeautifulSoup(r.text)
```

```python
allrow = soup.findAll('tr')
```

```python
el = allrow[1].findAll('td')
a = el[4].contents[-1]
for ch in ['$',',']:
    a = a.replace(ch,'')
a
```

```python
result = {}
yearlist = []
titlelist = []
dbo = []
ibo = []
wbo = []
result['Release year'] = yearlist
result['Title'] = titlelist
result['Domestic'] = dbo
result['International'] = ibo
result['Worldwide'] = wbo
```

```python
for row in allrow[1:101]:
    el = row.findAll('td')
    dbonumber = el[3].contents[-1].replace(',','').strip('$')
    ibonumber = el[4].contents[-1].replace(',','').strip('$')
    wbonumber = el[5].contents[-1].replace(',','').strip('$')
    yearlist.append(el[1].contents[-1])
    titlelist.append(el[2].contents[-1].contents[-1].contents[-1])
    dbo.append(dbonumber)
    ibo.append(ibonumber)
    wbo.append(wbonumber)
print(result)
    
```

```python
df = pd.DataFrame(result)
df.head(10)
```

```python
df = df[['Release year', 'Title', 'Domestic', 'International', 'Worldwide']]
df
```

```python
df.to_csv('moviestats.csv')
```
